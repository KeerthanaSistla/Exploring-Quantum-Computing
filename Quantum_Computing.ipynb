#Quantum Dice Simulator

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

simulator = AerSimulator()

def quantum_dice():
    while True:

        qc = QuantumCircuit(3, 3)

        # Put all qubits into superposition
        qc.h(0)
        qc.h(1)
        qc.h(2)

        # Measure
        qc.measure([0, 1, 2], [0, 1, 2])

        job = simulator.run(qc, shots=1)

        result = job.result()

        counts = result.get_counts()

        binary = list(counts.keys())[0]

        number = int(binary, 2)

        if 1 <= number <= 6:
            return number

print("🎲 Quantum Dice Roll:", quantum_dice())
```

```python
!pip install -q qiskit qiskit-aer

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from collections import Counter
import matplotlib.pyplot as plt

simulator = AerSimulator()

def quantum_dice():
    while True:
        qc = QuantumCircuit(3, 3)

        qc.h(0)
        qc.h(1)
        qc.h(2)

        qc.measure([0, 1, 2], [0, 1, 2])

        job = simulator.run(qc, shots=1)
        result = job.result()

        counts = result.get_counts()
        binary = list(counts.keys())[0]

        number = int(binary, 2)

        if 1 <= number <= 6:
            return number

print("=" * 50)
print("QUANTUM DICE SIMULATOR")
print("=" * 50)

roll = quantum_dice()

print(f"\n🎲 Quantum Dice Result: {roll}")

print("\nRolling 10 times...\n")

for i in range(10):
    print(f"Roll {i+1}: {quantum_dice()}")

qc = QuantumCircuit(3, 3)

qc.h(0)
qc.h(1)
qc.h(2)

qc.measure([0, 1, 2], [0, 1, 2])

print("\nQuantum Circuit:\n")
print(qc)

print("\nRunning 1000 quantum dice rolls...")

results = [quantum_dice() for _ in range(1000)]

frequency = Counter(results)

print("\nFrequency Distribution:\n")

for face in range(1, 7):
    print(f"Face {face}: {frequency[face]}")

faces = list(range(1, 7))
counts = [frequency[i] for i in faces]

plt.figure(figsize=(8, 5))
plt.bar(faces, counts)
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.title("Quantum Dice Distribution (1000 Rolls)")
plt.xticks(faces)
plt.show()
```

#Bell State (Quantum Entanglement)

```python
!pip install -q qiskit qiskit-aer matplotlib

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

qc = QuantumCircuit(2, 2)

qc.h(0)

qc.cx(0, 1)

qc.measure([0, 1], [0, 1])

print("Quantum Circuit:\n")
print(qc)

simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()

counts = result.get_counts()

print("\nMeasurement Results:\n")
print(counts)

plot_histogram(counts)
plt.show()

print("\nIndividual Measurements:")

job = simulator.run(qc, shots=10)

result = job.result()

counts = result.get_counts()

for state, frequency in counts.items():
    print(f"{state} -> {frequency}")
```

#Quantum Teleportation

```python
pip install qiskit qiskit-aer matplotlib
```

```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create quantum and classical registers
q = QuantumRegister(3, 'q')
c = ClassicalRegister(3, 'c')

qc = QuantumCircuit(q, c)

# ------------------------------------------------------------------
# Step 1: Prepare the state to teleport (|+>)
# ------------------------------------------------------------------
qc.h(q[0])

# ------------------------------------------------------------------
# Step 2: Create Bell pair between qubit 1 and qubit 2
# ------------------------------------------------------------------
qc.h(q[1])
qc.cx(q[1], q[2])

# ------------------------------------------------------------------
# Step 3: Bell-state measurement
# ------------------------------------------------------------------
qc.cx(q[0], q[1])
qc.h(q[0])

# Measure qubits 0 and 1
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])

# ------------------------------------------------------------------
# Step 4: Conditional corrections
# ------------------------------------------------------------------

# If c[1] == 1, apply X
with qc.if_test((c[1], 1)):
    qc.x(q[2])

# If c[0] == 1, apply Z
with qc.if_test((c[0], 1)):
    qc.z(q[2])

# ------------------------------------------------------------------
# Step 5: Measure teleported qubit
# ------------------------------------------------------------------
qc.measure(q[2], c[2])

# Display circuit
print("Quantum Teleportation Circuit:\n")
print(qc.draw())

# ------------------------------------------------------------------
# Simulation
# ------------------------------------------------------------------
simulator = AerSimulator()

compiled_circuit = transpile(qc, simulator)

result = simulator.run(compiled_circuit, shots=1000).result()

counts = result.get_counts()

print("\nResults:\n")
print(counts)

# Plot histogram
plot_histogram(counts)
plt.show()
```
